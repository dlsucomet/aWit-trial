```python
top10_names = ["danceability", "loudness", "speechiness", "liveness", "valence", "tempo", "duration"]
top10_mean = [top10["danceability"].mean(), top10["loudness"].mean(), top10["speechiness"].mean(), top10["liveness"].mean(), top10["valence"].mean(), top10["tempo"].mean(), top10["duration_ms"].mean()]
top10_median = [top10["danceability"].median(), top10["loudness"].median(), top10["speechiness"].median(), top10["liveness"].median(), top10["valence"].median(), top10["tempo"].median(), top10["duration_ms"].median()]
top10_min = [top10["danceability"].min(), top10["loudness"].min(), top10["speechiness"].min(), top10["liveness"].min(), top10["valence"].min(), top10["tempo"].min(), top10["duration_ms"].min()]
top10_max = [top10["danceability"].max(), top10["loudness"].max(), top10["speechiness"].max(), top10["liveness"].max(), top10["valence"].max(), top10["tempo"].max(), top10["duration_ms"].max()]
top10_std = [top10["danceability"].std(), top10["loudness"].std(), top10["speechiness"].std(), top10["liveness"].std(), top10["valence"].std(), top10["tempo"].std(), top10["duration_ms"].std()]

top10_stats = {'Summary Stat':top10_names, 'Mean':top10_mean, 'Median':top10_median, 'Standard Dev':top10_std, 'Minimum':top10_min, 'Maximum':top10_max}
top10 = pd.DataFrame(top10_stats)
top10
```
