```python
## rate of change (compares the week stream to the previous week stream)
```

```python
streams = stream.drop(columns=['artist_names'])
streams.set_index('track_name', inplace=True)
streams_change = streams.pct_change(axis=1)
streams_change
```

```python
streams_change = streams_change.mean(axis=1)
streams_change
```

```python
streams_change = streams_change.to_frame('streams mean')
streams_change_df = pd.DataFrame(streams_change)
streams_change_df
```

```python
streams_change_df = streams_change_df['streams mean'].sort_values(ascending=False)
streams_change_df
```
