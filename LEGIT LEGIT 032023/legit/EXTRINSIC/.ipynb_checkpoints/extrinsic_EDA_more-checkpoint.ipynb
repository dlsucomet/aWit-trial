```python
# are there more songs from well-known artists (remained for 73 consecutive weeks on Weekly Top Artists PH chart) in the Top 50 songs chart?
```

```python
songs_top_artists = extrinsic[extrinsic.chart_history == 73.0]
songs_top_artists
```

```python
# more songs from well-known artists (28 songs)
```

```python
top_artists = songs_top_artists['artist'].value_counts(ascending=False).to_frame()
top_artists = top_artists.reset_index()
top_artists.columns = ['artist', 'song count']
top_artists
```

```python
# are there not well-known artists who have more than 1 song in the chart?
```

```python
songs_other_artists = extrinsic[extrinsic.chart_history != 73.0]
other_artists = songs_other_artists['artist'].value_counts(ascending=False).to_frame()
other_artists = other_artists.reset_index()
other_artists.columns = ['artist', 'song count']
other_artists
```

```python
# what is the google trend for top artists?
```

```python
songs_top_artists = songs_top_artists.drop(columns=['track_name', 'artist_collab', 'chart_history', 'music_label'])
songs_top_artists = songs_top_artists.groupby('artist', as_index=False).first()
songs_top_artists = songs_top_artists.loc[:, ~songs_top_artists.columns.str.contains('^Unnamed')]
songs_top_artists
```

```python
top_artists_trend = songs_top_artists.T
top_artists_trend.columns = top_artists_trend.iloc[0] 
top_artists_trend = top_artists_trend[1:]
top_artists_trend.plot(ylabel='Google Trends Search Volume Index', xlabel='Days', figsize=(20,10))
```

```python
# only top artists with more than 1 song
```

```python
top_artists_trend = songs_top_artists.loc[songs_top_artists['artist'].isin(['sza','adie','moira dela torre','zack tabudlo','arthur nery','joji','chris brown','taylor swift'])]
top_artists_trend = top_artists_trend.T
top_artists_trend.columns = top_artists_trend.iloc[0] 
top_artists_trend = top_artists_trend[1:]
top_artists_trend.plot(ylabel='Google Trends Search Volume Index', xlabel='Days', figsize=(20,10))
```

```python
# trend for not well-known artists with more than 1 song
```

```python
songs_other_artists = extrinsic[extrinsic.chart_history != 73.0]
songs_other_artists = songs_other_artists.drop(columns=['track_name', 'artist_collab', 'chart_history', 'music_label'])
songs_other_artists = songs_other_artists.groupby('artist', as_index=False).first()
songs_other_artists = songs_other_artists.loc[:, ~songs_other_artists.columns.str.contains('^Unnamed')]
songs_other_artists
```

```python
other_artists_trend = songs_other_artists.loc[songs_other_artists['artist'].isin(['newjeans', 'stephen sanchez'])]
other_artists_trend = other_artists_trend.T
other_artists_trend.columns = other_artists_trend.iloc[0] 
other_artists_trend = other_artists_trend[1:]
other_artists_trend.plot(ylabel='Google Trends Search Volume Index', xlabel='Days', figsize=(20,10))
```

```python
temp1 = songs_other_artists.loc[songs_other_artists['artist'].isin(['newjeans'])]
temp2 = songs_top_artists.loc[songs_top_artists['artist'].isin(['sza', 'adie'])]
compare_trend = pd.concat([temp1, temp2], ignore_index = True)
compare_trend.reset_index()
compare_trend = compare_trend.T
compare_trend.columns = compare_trend.iloc[0] 
compare_trend = compare_trend[1:]
compare_trend.plot(ylabel='Google Trends Search Volume Index', xlabel='Days', figsize=(20,10))
```

```python
weeklyGoogleTrend = compare_trend.iloc[[6,13,20,27,34,41,48,55,62,69], :]
weeklyGoogleTrend.plot(ylabel='Google Trends Search Volume Index', xlabel='Weeks', figsize=(20,10))
```

```python
streams = pd.read_csv('10WeeksStream.csv')
streams = streams.loc[:, ~streams.columns.str.contains('^Unnamed')]
streams
```

```python
stream_artist = streams.loc[streams['artist_names'].isin(['SZA', 'Adie', 'NewJeans'])]
stream_artist = stream_artist.drop(columns=['track_name'])
stream_artist = stream_artist.T
stream_artist.columns = stream_artist.iloc[0] 
stream_artist = stream_artist[1:]
stream_artist.plot(ylabel='Streams', xlabel='Weeks', figsize=(20,10))
```
