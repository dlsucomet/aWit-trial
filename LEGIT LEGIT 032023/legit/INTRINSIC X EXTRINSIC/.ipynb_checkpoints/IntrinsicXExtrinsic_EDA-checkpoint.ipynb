```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
intrinsic = pd.read_csv('ThesisHits.csv')
intrinsic
```

```python
extrinsic = pd.read_csv('extrinsic.csv')
extrinsic
```

```python
merged = pd.concat([intrinsic, extrinsic], axis=1)
merged
```

```python
## intrinsic features stats based on music label
```

```python
## major
```

```python
major = merged[merged.music_label == 'Major']
major = major[['title', 'artist/s', 'danceability', 'key', 'loudness', 'speechiness', 'liveness', 'valence', 'tempo', 'duration_ms']]
major
```

```python
columnNames = ["danceability", "loudness", "speechiness", "liveness", "valence", "tempo", "duration"]
major_mean = [major["danceability"].mean(), major["loudness"].mean(), major["speechiness"].mean(), major["liveness"].mean(), major["valence"].mean(), major["tempo"].mean(), major["duration_ms"].mean()]
major_median = [major["danceability"].median(), major["loudness"].median(), major["speechiness"].median(), major["liveness"].median(), major["valence"].median(), major["tempo"].median(), major["duration_ms"].median()]
major_min = [major["danceability"].min(), major["loudness"].min(), major["speechiness"].min(), major["liveness"].min(), major["valence"].min(), major["tempo"].min(), major["duration_ms"].min()]
major_max = [major["danceability"].max(), major["loudness"].max(), major["speechiness"].max(), major["liveness"].max(), major["valence"].max(), major["tempo"].max(), major["duration_ms"].max()]
major_std = [major["danceability"].std(), major["loudness"].std(), major["speechiness"].std(), major["liveness"].std(), major["valence"].std(), major["tempo"].std(), major["duration_ms"].std()]

major_stats = {'Summary Stat':columnNames, 'Mean':major_mean, 'Median':major_median, 'Standard Dev':major_std, 'Minimum':major_min, 'Maximum':major_max}
major1 = pd.DataFrame(major_stats)
major1
```

```python
major_stats_2 = {'Summary Stat':["key"], 'Mode': [major["key"].mode()[0]], 'Median':[major["key"].median()], 'Standard Dev':[major["key"].std()], 'Minimum':[major["key"].min()], 'Maximum':[major["key"].max()]}
major_2 = pd.DataFrame(major_stats_2)
major_2
```
